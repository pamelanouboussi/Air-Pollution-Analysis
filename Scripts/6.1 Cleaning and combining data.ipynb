### Cleaning the non numerical data in the PM2.5 and 10 annual mean colums
- the ()-converted value" in WHO (World Health Organization) air quality data indicates that the reported air quality measurement has been converted from its original unit to a standardized unit for comparison and analysis purposes.
- the text around the values will be removed as they represent converted values. They were converted to PM2.5 or PM10

```python
# created new colum to place new data in
# use str.replace() to remove any non-digit characters from the "PM2.5 Annual mean" column
air_quality['PM2.5 Annual mean cleaned'] = air_quality['PM2.5 Annual mean'].str.replace(r'\D', '', regex=True)
```

```python
#check cleaned colum to see if there are any non numeric values
non_numeric_pm25_a = pd.to_numeric(air_quality['PM2.5 Annual mean cleaned'], errors='coerce').isnull().sum()
non_numeric_pm25_a
```

```python
# created new colum to place new data in
# use str.replace() to remove any non-digit characters from the "PM2.5 Annual mean" column
air_quality['PM10 Annual mean cleaned'] = air_quality['PM10 Annual mean'].str.replace(r'\D', '', regex=True)
```

```python
#check cleaned colum to see if there are any non numeric values
non_numeric_pm10_a = pd.to_numeric(air_quality['PM10 Annual mean cleaned'], errors='coerce').isnull().sum()
non_numeric_pm10_a
```
